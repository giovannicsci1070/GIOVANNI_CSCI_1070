```python

class_labels = LabelEncoder()
```

```python
titanic_df_new = titanic_df.dropna()
titanic_df_new.head(20)
```

```python
Sex_map = {'male':1,'female':2}
titanic_df['Sex'] = titanic_df['Sex'].map(Sex_map)


titanic_df_new['Sex'] = titanic_df_new['Sex'].map(Sex_map)


titanic_df_new['ticket'] = class_labels.fit_transform(titanic_df_new['Ticket'].values)
titanic_df_new['cabin'] = class_labels.fit_transform(titanic_df_new['Cabin'].values)
titanic_df_new['embarked'] = class_labels.fit_transform(titanic_df_new['Embarked'].values)
titanic_df_new['fare'] = class_labels.fit_transform(titanic_df_new['Fare'].values)
titanic_df_new['name'] = class_labels.fit_transform(titanic_df_new['Name'].values)


titanic_df_new = titanic_df_new.drop('Ticket',axis=1)
titanic_df_new = titanic_df_new.drop('Name',axis=1)
titanic_df_new = titanic_df_new.drop('Cabin',axis=1)
titanic_df_new = titanic_df_new.drop('Fare',axis=1)
titanic_df_new = titanic_df_new.drop('Embarked',axis=1)

titanic_df_new.head(20)
```

```python
correlation = titanic_df_new.corr()
sns.heatmap(correlation, annot=True)
```

```python
titanic_df_new = titanic_df_new.drop('ticket',axis=1)
titanic_df_new = titanic_df_new.drop('Pclass',axis=1)
titanic_df_new = titanic_df_new.drop('cabin',axis=1)
titanic_df_new = titanic_df_new.drop('Parch',axis=1)
```
