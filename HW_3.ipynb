```python
#Starter code. Do NOT modify this cell
import pandas as pd
import numpy as np

# Read in dataset
df = pd.read_csv('cars.csv',delimiter=';')

# Convert each column to an np.array
car = df.Car.to_numpy()
mpg = df.MPG.to_numpy()
cylinders = df.Cylinders.to_numpy()
displacement = df.Displacement.to_numpy()
horsepower = df.Horsepower.to_numpy()
weight = df.Weight.to_numpy()
acceleration = df.Acceleration.to_numpy()
model = df.Model.to_numpy()
origin = df.Origin.to_numpy()

# create an array of headers
headers = np.array(['Car', 'MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight','Acceleration', 'Model', 'Origin'])
```

```python
europe_cars_horsepower = horsepower[origin == 'Europe']
average_horsepower_europe = np.mean(europe_cars_horsepower)
print(average_horsepower_europe)
```

```python
average_mpg_japanese = np.mean(mpg[origin == 'Japan'])
average_mpg_us = np.mean(mpg[origin == 'US'])

print("Average mpg of Japanese cars:", average_mpg_japanese)
print("Average mpg of US cars:", average_mpg_us)
```

```python
correlation_coefficient = np.corrcoef(mpg, weight)[0, 1]

print("Correlation Coefficient between MPG and Weight:", correlation_coefficient)

# correlation coefficient is a way of measureing the relation of one variable to another
# values range between -1 to 1, at 0 there is nop correlatioan 
#  at -1 there is a negative correlation and as one variable increases the other will decrease

# in this case the correlation effecient is a negative value of -0.7875 
# as weight increases mpg decreases 
```
